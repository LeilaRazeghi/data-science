```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
seaborn.set()
data = pd.read_csv("california_cities.csv")
data.head()
```

```python
plt.scatter(data["longd"] , data["latd"] , s=data["area_total_km2"] , alpha=0.5 , linewidths=0 , c=np.log10(data["population_total"]) ,cmap='hsv' )
#map of cities 

plt.colorbar(label='log$_{10}$(population)')
plt.title("Area and population of California")
plt.xlabel("latitude")
plt.ylabel("longitude")
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')

plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='city area')

plt.show()
```
