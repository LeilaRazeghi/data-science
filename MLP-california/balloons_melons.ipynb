```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
b_widths = np.random.normal(20, 3, size=500)
b_lenghts = np.random.normal(30, 3, size=500)
b_weights = abs(np.random.normal(0.005, 0.01, size=500))

m_widths = np.random.normal(20, 2, size=500)
m_lengths = np.random.normal(20, 2, size=500)
m_weights = abs(np.random.normal(3, 0.5, size=500))

fig = plt.figure(figsize = (10, 9))
ax = plt.axes(projection = '3d')
ax.grid()
ax.scatter(xs = b_widths, ys = b_lenghts ,  zs = b_weights , color= 'red'    , marker="*" , label='balloons')
ax.scatter(xs = m_widths, ys = m_lengths, zs = m_weights, color= 'blue', marker = 'o', label = 'melons'  )


plt.title('balloons melons')
plt.xlabel('Widths(cm)')
plt.ylabel('Lengths(cm)')
ax.set_zlabel('weights(kg)', labelpad=-5)
plt.legend()
plt.show()
```
